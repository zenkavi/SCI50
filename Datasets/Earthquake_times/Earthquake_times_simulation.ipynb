```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

# Set the rate (lambda) of earthquakes per unit time
# For example, if on average one earthquake occurs every 10 days, then lambda = 1/10
lambda_rate = 1 / 10  # earthquakes per day

# Number of data points to simulate
n = 40

# Generate exponential data
inter_event_times = np.random.exponential(scale=1/lambda_rate, size=n)

# Print the simulated times
print("Simulated time between earthquakes (in days):")
print(inter_event_times)
```

```python
# Optional: plot a histogram to visualize the distribution
plt.hist(inter_event_times, bins=10, color='skyblue', edgecolor='black')
plt.title('Time Between Earthquakes')
plt.xlabel('Time (days)')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
import pandas as pd
```

```python
# Convert to DataFrame
df = pd.DataFrame({'inter_event_time_days': inter_event_times})
df.head()
```
